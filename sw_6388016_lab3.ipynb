```python
from sympy import solve
from sympy.plotting import plot
x = Symbol('x')
expr = 2*x**2 + 5*x - 3
solve(expr, dict=True) 
plot(expr)
```

```python
from sympy import *
from sympy.plotting import plot
x,y =symbols('x,y')
expr1 = 2*x + 4*y + 6
expr2 = 2*x + 3*y + 4
solve((expr1, expr2), dict=True)
pexp1= (-2*x-6)/4
pexp2 = (-2*x -4)/3
plot(pexp1, pexp2)
```
